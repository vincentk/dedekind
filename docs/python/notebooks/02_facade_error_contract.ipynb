# Dedekind Python Facade: Error Contract
This notebook checks that invalid payloads are rejected with `TypeError`.

```python
import dedekind
```

```python
def expect_type_error(label, fn):
    try:
        fn()
    except TypeError:
        print(f"{label}: ok")
        return
    raise AssertionError(f"{label}: expected TypeError")
```

```python
expect_type_error("ordered non-integral", lambda: dedekind.ordered_set_roundtrip([1, "x", 3]))
expect_type_error("unordered non-iterable", lambda: dedekind.unordered_set_roundtrip(42))
expect_type_error("path non-integral", lambda: dedekind.path_from_range([1, None, 3]))
print("notebook-02-ok")
```
