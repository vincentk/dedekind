# Dedekind Python Facade: Roundtrip Basics
This notebook validates the three MVP facade calls with deterministic inputs.

```python
import dedekind

ordered = dedekind.ordered_set_roundtrip([3, 1, 2, 2])
unordered = dedekind.unordered_set_roundtrip([4, 2, 4, 1])
path = dedekind.path_from_range([2, 4, 6, 8])

print(ordered)
print(unordered)
print(path)
```

```python
assert ordered == [1, 2, 3]
assert unordered == [1, 2, 4]
assert path == [2, 4, 6, 8]
print("notebook-01-ok")
```
