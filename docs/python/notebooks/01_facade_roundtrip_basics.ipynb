# Dedekind Python Facade: Happy Path
This notebook runs the MVP facade on valid inputs, prints the returned values, and asserts the expected results.

## Payload
Inputs: ordered `[3, 1, 2, 2]`, unordered `[4, 2, 4, 1]`, path `[2, 4, 6, 8]`.
Expected outputs: `[1, 2, 3]`, `[1, 2, 4]`, and `[2, 4, 6, 8]`.

```python
from pathlib import Path

def find_repo_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / "pyproject.toml").exists():
            return candidate
    raise FileNotFoundError("Could not locate repo root containing pyproject.toml")

repo_root = find_repo_root(Path.cwd())

try:
    import dedekind
except ModuleNotFoundError as err:
    raise ModuleNotFoundError(
        "dedekind is not installed in this notebook kernel. "
        f"From {repo_root}, run `python -m pip install --no-build-isolation -e .` "
        "or use `make integration-test` for the CI-style execution path."
    ) from err

print("happy-path: running facade calls")

ordered = dedekind.ordered_set_roundtrip([3, 1, 2, 2])
unordered = dedekind.unordered_set_roundtrip([4, 2, 4, 1])
path = dedekind.path_from_range([2, 4, 6, 8])

results = {
    "ordered_set_roundtrip": ordered,
    "unordered_set_roundtrip": unordered,
    "path_from_range": path,
}

for label, value in results.items():
    print(f"{label}: {value}")
```

```python
assert ordered == [1, 2, 3]
assert unordered == [1, 2, 4]
assert path == [2, 4, 6, 8]

print("happy-path: assertions passed")
print("notebook-01-ok")
```
